```python
dataset = []
for n in range (1,5):
    print(f"début page{n}")
    url = f"https://recherche-entreprises.api.gouv.fr/near_point?lat=45.182081&long=5.724301&radius=30&page={n}&per_page=[10]"
    response = requests.get(url=url)

    datas =response.json()

    for entreprise in datas['results'][:10]:  # Parcours direct, plus lisible
        nom = entreprise["nom_complet"]
        etablissements = entreprise["matching_etablissements"]

        for etablissement in etablissements:
            ligne = {
                "nom_entreprise": nom,
                "adresse": etablissement.get("adresse", ""),
                "latitude": etablissement.get("latitude", None),
                "longitude": etablissement.get("longitude", None),
                "activite_principale": etablissement.get("activite_principale", "")
            }
            dataset.append(ligne)

    print(f"page{n} faite")
    time.sleep(0.5)
```
